```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
```

```python
train = pd.read_csv('loan_train.csv')
test = pd.read_csv('loan_test.csv')
```

```python
categorical = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Credit_History']

for var in categorical:
    train = pd.concat([train, pd.get_dummies(train[var], prefix=var)], axis=1)
    del train[var]
```

```python
categorical = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Credit_History']

for var in categorical:
    test = pd.concat([test, pd.get_dummies(test[var], prefix=var)], axis=1)
    del test[var]
```

```python
continuous = ['Dependents','ApplicantIncome', 'CoapplicantIncome', 'LoanAmount','Loan_Amount_Term']

scaler = StandardScaler()

for var in continuous:
    train[var] = train[var].astype('float64')
    train[var] = scaler.fit_transform(train[var].values.reshape(-1, 1))
    
train.head()
```

```python
continuous = ['Dependents','ApplicantIncome', 'CoapplicantIncome', 'LoanAmount','Loan_Amount_Term']

scaler = StandardScaler()

for var in continuous:
    test[var] = test[var].astype('float64')
    test[var] = scaler.fit_transform(test[var].values.reshape(-1, 1))
    
test.head()
```

```python
train.isnull().sum()
```

```python
x_train = train.drop(['Loan_Status','Loan_ID'],axis=1)
y_train = train['Loan_Status']
```

```python
x_test = test.drop(['Loan_Status','Loan_ID'],axis=1)
y_test = test['Loan_Status']
```

```python
regression = LogisticRegression()
```

```python
regression.fit(x_train, y_train)
```

```python
accuracy = regression.score(x_test, y_test)
print('Test accuracy is:',accuracy)
```

```python
test['Prob'] = regression.predict_proba(x_test)[:,1]
test['Pred'] = regression.predict(x_test)
```
